```python
loss_train_arrs = np.squeeze(np.array(loss_train_arrs))
loss_valid_arrs = np.squeeze(np.array(loss_valid_arrs))
print(loss_train_arrs.shape, loss_valid_arrs.shape)
```

```python
# Create a 3x3 scatter plot matrix using matplotlib
fig, axes = plt.subplots(4, 2, figsize=(12, 12))

# Define the range for xlim and ylim
x_min, x_max = None, None
y_min, y_max = 0, 0.8

for i, ax in enumerate(axes.flat):
    if i < len(methods_style_train):
        ax.plot(loss_train_arrs[i], c=methods_style_train[i][1],linestyle=methods_style_train[i][2],marker=methods_style_train[i][3],markevery=20,markersize=5,
                label=methods_style_train[i][0])
        ax.plot(loss_valid_arrs[i], c=methods_style_valid[i][1],linestyle=methods_style_valid[i][2],marker=methods_style_valid[i][3],markevery=20,markersize=5,
                label=methods_style_valid[i][0])
    
        ax.set_xlabel('epoch')
        ax.set_ylabel('RMSE')
        # Set the same range for xlim and ylim
        ax.set_xlim(x_min, x_max)
        ax.set_ylim(y_min, y_max)
        ax.legend()
    else:
        ax.set_visible(False)
fig.suptitle('The convergence plots during training.', fontsize=16)
# Customize the plot appearance
plt.tight_layout()
plt.savefig(exp_dir + '{}_{}_convergence'.format(b_id,out_name), dpi=400)
plt.show()
```
