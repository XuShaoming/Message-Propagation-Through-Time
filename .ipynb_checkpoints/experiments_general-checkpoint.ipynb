```python
def MSE (Y, Yhat, mask = None):
    if type(Y) == torch.Tensor:
        if mask is None:
            mask = torch.ones(Y.size())
            mask.to(Y.device)
        return torch.sum(torch.mul(torch.square(Y-Yhat), mask)) / (torch.sum(mask) + 1)
    
    if mask is None:
        mask = np.ones(Y.shape)
    return np.sum(np.multiply(np.square(Y-Yhat), mask)) / (np.sum(mask) + 1)
```

```python
'''
This List contains the names of investigated variables.
['Date', 'PRECIPmm', 'TMP_MXdgC', 'TMP_MNdgC', 'SOLARMJ/m2', 'WINDM/s',
       'RHmd', 'SNOFALLmm', 'SNOMELTmm', 'SNOmm', 'PETmm', 'ETmm', 'SW_INITmm',
       'SW_ENDmm', 'PERCmm', 'GW_RCHGmm', 'DA_RCHGmm', 'REVAPmm', 'SA_STmm',
       'DA_STmm', 'SURQ_GENmm', 'SURQ_CNTmm', 'TLOSSmm', 'LATQGENmm',
       'LATQCNTmm', 'GW_Qmm', 'GW_Q_Dmm', 'DAILYCN', 'LAI', 'QTILEmm',
       'Q_pred_mm']
       
check: df['Date'][72378],  df['Date'][73050]
df['Date'][72378] == '2100-03-01'
df['Date'][73050] == '2102-01-02'
Both dates look fine. 
'''
def load_data(df, output_vars, input_vars):
    ''' load data and sperate data into features and labels.
    parameters
        df: a pandas data frame
        output_vars: A list contains the names of output variables.
        input_vars: A list contains the names of input variables. 
    Return:
        feat: an input data matrix
        labels: an output data matrix
    '''
    N = df.shape[0]
    feat = np.zeros((N,input_size))
    # Convert the dates to day of year(DOY), then bend the DOY at day 183 to 
    # give DOY a seasonal pattern as shwon in figure 
    dates = df['Date'].values
    doy = np.array([datetime.strptime(cdate, '%Y-%m-%d').date().timetuple().tm_yday for cdate in dates])
    doy = 183 - np.abs(doy - 183)
    # Load features
    feat = df[input_vars].values
    feat[:,0] = doy * 1.0

    labels = np.zeros((N,len(output_vars)))
    for i, var in enumerate(output_vars):
        if var in df.keys():
            labels[:,i] = df[var]
        elif var == 'SW_Delta':
            labels[:,i] = df['SW_ENDmm'].values - df['SW_INITmm'].values
        else:
            raise ValueError('Unexpected output variables {}'.format(var))
    
    return feat, labels
```

```python
# train_idx_arr = SampleData(train_idx,n_steps,shift_train,num_samples_train)
def SampleData(train_idx,n_steps,shift,num_samples):
    ''' SampleData a collection of index arrays. These index arrays can be applied on data to get data samples. 
    train_idx: a index array of data
    n_steps: sequence length (aka window size)
    shift: shift length.
        shift>0 : regular shift sampling
        shift=0 : random sampling
    num_samples: Is activated when shift=0. It determines the number of random samples.
    
    '''
    tN = train_idx.shape[0]
    print(tN)
    if shift>0:
        for i in np.arange(0,tN, shift):
            if i + n_steps > tN:
                break
            cur_idx = np.expand_dims(train_idx[i:i+n_steps],axis=0)
            if i == 0:
                train_idx_arr = cur_idx
            else:
                train_idx_arr = np.concatenate((train_idx_arr,cur_idx),axis=0)

        cur_idx = np.expand_dims(train_idx[-n_steps:],axis=0)
        train_idx_arr = np.concatenate((train_idx_arr,cur_idx),axis=0)

    if shift==0:
        start_idx = np.arange(0,tN-n_steps)
        start_idx = np.random.permutation(start_idx)
        num_samples = min([num_samples,start_idx.shape[0]])
        train_idx_arr = np.zeros((num_samples,n_steps)).astype(int)
        for i in range(num_samples):
            train_idx_arr[i,:] = train_idx[start_idx[i]:start_idx[i]+n_steps]
            
    return train_idx_arr
```

```python
def get_rmses_itr(y, yhats):
    rmses = []
    for yhat in yhats:
        rmses.append(np.sqrt(MSE(y, yhat)))
    return np.asarray(rmses)
```

```python
def get_NSE(Y, Y_hat):
    return 1 - np.sum(np.square(Y-Y_hat)) / np.sum(np.square(Y - np.mean(Y)))
```

```python
def get_NSE_itr(y,yhats):
    NSEs = []
    for yhat in yhats:
        NSEs.append(get_NSE(y, yhat))
    return np.asarray(NSEs)
```

```python
def get_step_mse(y_sp, pred_itr):
    mses = []
    for j in range(366):
        mses.append(MSE(pred_itr[:,j], y_sp[:,j]))
    return np.squeeze(mses)

def get_step_mse_itrs(y_sp, pred_itrs):
    step_mse_iters = []
    for pred_itr in pred_itrs:
        step_mse_iters.append(get_step_mse(y_sp, pred_itr))
    return np.squeeze(step_mse_iters)
```

```python
def get_agg_rmses_itrs(pred_agg_itrs, y, max_iter, n_classes):
    rmses = np.zeros((max_iter, n_classes))
    for i in range(max_iter):
        for j in range(n_classes):
            rmses[i,j] = np.sqrt(MSE(pred_agg_itrs[i][:,j], y))
    return rmses
```

```python
def get_NSE(Y, Y_hat):
    return 1 - np.sum(np.square(Y-Y_hat)) / np.sum(np.square(Y - np.mean(Y)))
```

```python
def get_NSE_itr(y,yhats):
    NSEs = []
    for yhat in yhats:
        NSEs.append(get_NSE(y, yhat))
    return np.asarray(NSEs)
```

```python
rmses = []
for pred_itrs in predictions_agg:
    rmses.append(get_agg_rmses_itrs(pred_itrs, np.squeeze(test_y),max_iter=run_iter, n_classes=1))
rmses = np.squeeze(rmses)
```

```python
nses = []
for pred_itrs in predictions_agg:
    nses.append(get_NSE_itr(np.squeeze(test_y),np.squeeze(pred_itrs)))
nses = np.squeeze(nses)
```

## Analysis daily RMSE

```python
dates = df['Date'].values
doys = np.array([datetime.strptime(cdate, '%Y-%m-%d').date().timetuple().tm_yday for cdate in dates])
output_vars = np.asarray(output_vars)

idx = np.where(output_vars == output_var)[0][0]
doy = doys[train_len+valid_len:]
doy = doy[366*2:]
```

```python
y = test_y[:,idx]
daily_rmses_methods = []
for i, agg_itr in enumerate(predictions_agg):
    daily_rmses_list = []
    for preds_aggs in agg_itr:
        preds_agg = preds_aggs[:,idx]
        # print(preds_agg.shape)
        # print(y.shape)
        df_res = pd.DataFrame(data = {'doy':doy, 'pred':preds_agg, 'obs':y})
        daily_rmses = []
        assert(min(df_res['doy'])==1 and max(df_res['doy'])==366)
        for j in range(1, 367):
            chunk = df_res[df_res['doy'] == j]
            daily_rmses.append(np.sqrt(MSE(chunk['pred'], chunk['obs'])))
        daily_rmses_list.append(np.squeeze(daily_rmses))
    daily_rmses_methods.append(daily_rmses_list)
```

```python
daily_rmses_methods = np.squeeze(daily_rmses_methods)
daily_rmses_methods.shape
daily_rmses_methods_mean = np.mean(daily_rmses_methods, axis=1)
daily_rmses_methods_std = np.std(daily_rmses_methods, axis=1)
```
