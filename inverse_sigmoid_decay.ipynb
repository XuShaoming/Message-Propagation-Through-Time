```python
import numpy as np
import matplotlib.pyplot as plt
import math
np.random.seed(4)
```

```python
def inverse_sigmoid_decay(epoch, epochs, alpha, beta):
    return 1 / (1 + np.exp(alpha*(epoch/epochs - beta)))
epochs = 400
alpha=10
beta=1/2
print(inverse_sigmoid_decay(0, epochs, alpha, beta))
print(inverse_sigmoid_decay(200, epochs, alpha, beta))
print(inverse_sigmoid_decay(400, epochs, alpha, beta))
decay = [inverse_sigmoid_decay(epoch, epochs, alpha, beta) for epoch in range(400)]
decay = decay + list(np.zeros(100))

caravan_epochs = 150
caravan_alpha=10
caravan_beta=1/2
print(inverse_sigmoid_decay(0, caravan_epochs, caravan_alpha, caravan_beta))
print(inverse_sigmoid_decay(100, caravan_epochs, caravan_alpha, caravan_beta))
print(inverse_sigmoid_decay(200, caravan_epochs, caravan_alpha, caravan_beta))
caravan_decay = [inverse_sigmoid_decay(epoch, caravan_epochs, caravan_alpha, caravan_beta) for epoch in range(caravan_epochs)]
caravan_decay = caravan_decay + list(np.zeros(50))

ig,ax = plt.subplots(1,1, figsize=(10,3))

ax.plot(range(500),decay,c='b', label='Synthetic data experiments')
ax.plot(range(200),caravan_decay,c='r', label='Real world data experiments')
plt.xlabel("Epochs")
plt.ylabel("Decay Rate")
plt.legend()
plt.title("Rates of Inverse Sigmoid Decay in Scheduled Sampling")
plt.tight_layout()
plt.show()
plt.close()
```
